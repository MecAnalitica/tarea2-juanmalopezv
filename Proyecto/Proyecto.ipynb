## Mapeo Logistico

Ecuacion logistica
\begin{equation}
x_{n+1}=rx_{n}(1-x_{n})
\end{equation}

donde **r** es la taza de crecimiento
$x_{n}$ la poblacion en la n temporada
$x_{n+1}$ la pobalacion en la n+1 temporada


```python
from math import *
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.zeros(201)
y=np.zeros(201)
z=np.zeros(201)

x2=np.zeros(201)
y2=np.zeros(201)
z2=np.zeros(201)
```

```python
x0=0.5
r=3.9
i=0 

while i<200:
    x1=r*x0*(1-x0)
    i+=1
    x[i]=i
    y[i]=x0
    z[i]=x1
    x0=x1
 
r=3.8
i=0 

while i<200:
    x1=r*x0*(1-x0)
    i+=1
    x2[i]=i
    y2[i]=x0
    z2[i]=x1
    x0=x1
```

```python
plt.xlim(0,25)
plt.plot(x, y, color='g')
plt.plot(x2, y2)
plt.xlabel("Time")
plt.ylabel("Population")
plt.legend("r=3.9","r=3.8")
plt.grid()
plt.show()
```

## Exponente de Lyapunov

```python
from math import *
import random 
```

```python
def expl (n, A):
    lya=[]
    l=0
    for i in range(len(A)):
        x=0
        l=0
        x0=random.random()
        for j in range(n):
            x=A[i]*x0*(1-x0)
            x0=x
            l=l+log(abs(A[i]-2*A[i]*x))
        lya.append(l/n)
    return lya
```

```python
n=10000
r=np.linspace(3,4, 1000)
lyp=expl (n, r)
```

```python
plt.plot(r, lyp, '-')
plt.xlabel('r')
plt.ylabel('$\lambda$')
plt.axhline(0, color='g')
plt.grid()
plt.show()
```

## Stranger Attractors

```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
from matplotlib import style
from scipy.integrate import odeint
```

Atractor de Loretz

\begin{equation}
\frac{dx}{dt}=a(y-x)\\
\frac{dy}{dt}=x(b-z)-y\\
\frac{dz}{dt}=xy-cz
\end{equation}

```python
def lorentz(w, t):
    a=10
    b=28
    c=8/3
    
    x=w[0]
    y=w[1]
    z=w[2]
    
    dx=a*(y-x)
    dy=(x*(b-z))-y
    dz=(x*y)-(c*z)
    
    dw=[dx,dy,dz]
    return dw

w0=[1,0.1,1]

t=np.linspace(0,12,10000)
w=odeint(lorentz,w0,t)

x=w[:,0]
y=w[:,1]
z=w[:,2]

fig = plt.figure()
plane = fig.add_subplot(111,projection='3d')
plt.plot(x, y, z, ',', color='blue')
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

Atractor de Rossler

Las ecuaciones del sistema son las siguientes

\begin{equation}
\frac{dx}{dt}=-y-z\\
\frac{dy}{dt}=x+ay\\
\frac{dz}{dt}=b+z(x-c)
\end{equation}

```python
def rossler(w, t):
    a=0.2
    b=0.2
    c=5.7
    
    x=w[0]
    y=w[1]
    z=w[2]
    
    dx=-y-z
    dy=x+a*y
    dz=b+z*(x-c)
    
    dw=[dx,dy,dz]
    return dw

w0=[-0.5,1,1]

t=np.linspace(0,300,100000)
w=odeint(rossler,w0,t)

x=w[:,0]
y=w[:,1]
z=w[:,2]

fig = plt.figure()
plane = fig.add_subplot(111,projection='3d')
plt.plot(x, y, z, ',', color='purple')
plt.xlabel('x')
plt.ylabel('y')


plt.show()
```

Atractor de Thomas

\begin{equation}
\frac{dx}{dt}=sin(y)-bx\\
\frac{dy}{dt}=sin(z)-by\\
\frac{dz}{dt}=sin(x)-bz
\end{equation}

```python
def thomas(w, t):
    b=0.208
    
    x=w[0]
    y=w[1]
    z=w[2]
    
    dx=np.sin(y)-b*x
    dy=np.sin(z)-b*y
    dz=np.sin(x)-b*z
    
    dw=[dx,dy,dz]
    return dw

w0=[3,1,4]

t=np.linspace(0,300,10000)
w=odeint(thomas,w0,t)

x=w[:,0]
y=w[:,1]
z=w[:,2]

fig = plt.figure()
plane = fig.add_subplot(111,projection='3d')
plt.plot(x, y, z, ',', color='brown')
plt.xlabel('x')
plt.ylabel('y')


plt.show()
```

Atractor de Halvorsen

\begin{equation}
\frac{dx}{dt}=-ax+2y-4z-y^{3}+(3a+15)\\
\frac{dy}{dt}=-ay+2z-4x-z^{2}+(3a+15)\\
\frac{dz}{dt}=-az+2x-4y-x^{2}+(3a+15)
\end{equation}

```python
def halvorsen(w, t):
    a=1.4
    
    x=w[0]
    y=w[1]
    z=w[2]
    
    dx=-(a*x)-(4*y)-(4*z)-(y**2)
    dy=-(a*y)-(4*z)-(4*x)-(z**2)
    dz=-(a*z)-(4*x)-(4*y)-(x**2)
    
    dw=[dx,dy,dz]
    return dw

w0=[0.1,1,0.2]

t=np.linspace(0,50,10000)
w=odeint(halvorsen,w0,t)

x=w[:,0]
y=w[:,1]
z=w[:,2]

fig = plt.figure()
plane = fig.add_subplot(111,projection='3d')
plt.plot(x, y, z, ',', color='green')

plt.xlabel('x')
plt.ylabel('y')


plt.show()
```

## Fractales

Conjunto de Maldelbrot

\begin{equation}
z_{n+1}=z^{2}_{n}+c
\end{equation}

```python
from pylab import imshow,show,gray
from numpy import zeros,linspace

n=800

M = zeros([n,n],int)
xvalues = linspace(-2,2,n)
yvalues = linspace(-2,2,n)

for u,x in enumerate(xvalues):
    for v,y in enumerate(yvalues):
        z = 0 + 0j
        c = complex(x,y)
        for i in range(100):
            z = z*z + c
            if abs(z) > 2.0:
                M[v,u] = 1
                break

imshow(M,origin="lower")
gray()
show()
```

## Modelo Lotka Volterra

Ecuaciones Lotka Volterra

\begin{equation}
\frac{dx}{dt}=ax-bxy\\
\frac{dy}{dt}=cxy-dy
\end{equation}

```python
def lotkav(w, t):
    a=8
    b=5
    c=3
    d=1
    
    x=w[0]
    y=w[1]

    
    dx=a*x-b*x*y
    dy=d*x*y-c*y

    
    dw=[dx,dy]
    return dw

w0=[1,0.1]

t=np.linspace(0,12,5000)
w=odeint(lotkav,w0,t)

x=w[:,0]
y=w[:,1]

plt.plot(t, y, color='red')
plt.plot(t, x, ',', color='g')
plt.legend(('Predador', 'Prey'))
plt.xlabel('time')
plt.ylabel('Population')

plt.show()
```
